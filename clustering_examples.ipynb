# Clustering Examples

This notebook contains some code you may use to supplement the <code>assignment_2.ipynb</code> notebook. 

```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import Birch
from sklearn.cluster import MeanShift
from sklearn.mixture import GaussianMixture
```

## Generate Data

```python
data, class_labels = make_classification(n_samples=3000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=2, random_state=4)
total_classes = len(np.unique(class_labels))
```

```python
# create scatter plot for samples from each class
for class_value in range(total_classes):
    # get row indexes for samples with this class
    row_ix = np.where(class_labels == class_value)
    # create scatter of these samples
    plt.scatter(data[row_ix, 0], data[row_ix, 1])
# show the plot
plt.show()
```

## Affinity Propagation

```python
# define the model
model = AffinityPropagation(damping=0.8)

# fit the model
model.fit(data)

# assign a cluster to each example
clusters = model.predict(data)

# retrieve unique clusters
uniq_clusters = np.unique(clusters)
print(f'Found {len(uniq_clusters)} clusters')

# create scatter plot for samples from each cluster
for cluster in uniq_clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(clusters == cluster)
    # create scatter of these samples
    plt.scatter(data[row_ix, 0], data[row_ix, 1])
# show the plot
plt.show()
```

### Agglomerative Clustering

```python
# define the model
model = AgglomerativeClustering(n_clusters=2)

# fit model and predict clusters
clusters = model.fit_predict(data)

# retrieve unique clusters
uniq_clusters = np.unique(clusters)
print(f'Found {len(uniq_clusters)} clusters')

# create scatter plot for samples from each cluster
for cluster in uniq_clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(clusters == cluster)
    # create scatter of these samples
    plt.scatter(data[row_ix, 0], data[row_ix, 1])

plt.show()
```

### BIRCH

```python
# define the model
model = Birch(threshold=0.01, n_clusters=3)

# fit the model
model.fit(data)

# assign a cluster to each example
clusters = model.predict(data)

# retrieve unique clusters
uniq_clusters = np.unique(clusters)
print(f'Found {len(uniq_clusters)} clusters')

# create scatter plot for samples from each cluster
for cluster in uniq_clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(clusters == cluster)
    # create scatter of these samples
    plt.scatter(data[row_ix, 0], data[row_ix, 1])

plt.show()
```

## Mean Shift

```python
# define the model
model = MeanShift()

# fit model and predict clusters
clusters = model.fit_predict(data)

# retrieve unique clusters
uniq_clusters = np.unique(clusters)
print(f'Found {len(uniq_clusters)} clusters')

# create scatter plot for samples from each cluster
for cluster in uniq_clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(clusters == cluster)
    # create scatter of these samples
    plt.scatter(data[row_ix, 0], data[row_ix, 1])

plt.show()
```

## Gaussian Mixture Model

```python
# define the model
model = GaussianMixture(n_components=4)

# fit the model
model.fit(data)

# assign a cluster to each example
clusters = model.predict(data)

# retrieve unique clusters
uniq_clusters = np.unique(clusters)
print(f'Found {len(uniq_clusters)} clusters')

# create scatter plot for samples from each cluster
for cluster in uniq_clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(clusters == cluster)
    # create scatter of these samples
    plt.scatter(data[row_ix, 0], data[row_ix, 1])

plt.show()
```
